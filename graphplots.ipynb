```python
class graphplots:
    def graph(x,number_of_columns,newpath):
        import pandas as pd
        import numpy as np
        import matplotlib.pyplot as plt
        import seaborn as sns    
        import os

        cars=pd.read_csv('mtcars.csv')

        if not os.path.exists(newpath):
            os.makedirs(newpath)
        #dir_name = input("Please a enter:")
        if number_of_columns == None:
            for i, col in enumerate(cars.columns):
                if len(cars[col].unique())<5:
                    sns.barplot(x=cars[col],y=cars.index) 
                    plt.title('Barplot of {0}'.format(col))
                    plt.xlabel('{0}'.format(col))
                    plt.ylabel('Count of cars')
                    plt.savefig(newpath+'//barplot_{1}.png'.format(col))
                    plt.show()

                else:
                    plt.hist(cars[col]) 
                    plt.title('Histogram of {0}'.format(col))
                    plt.xlabel('{0}'.format(col))
                    plt.ylabel('Frequecy of {0}'.format(col))
                    plt.savefig(newpath+'//histogram_{0}.png'.format(col))
                    plt.show()
                    plt.boxplot(cars[col]) 
                    plt.title('Boxplot of {0}'.format(col))
                    plt.xlabel('{0}'.format(col))
                    plt.savefig(newpath+'//boxplot_{0}.png'.format(col))
                    plt.show()
        else:
            all_columns_we_need = cars.columns[number_of_columns]
            needed_dataframe = cars[all_columns_we_need]

             #cars=pd.read_csv('mtcars.csv')
                #our_columns = needed_dataframe[[1,3,4]]
            for i, col in enumerate(needed_dataframe.columns):
                if len(needed_dataframe[col].unique())<5:
                    sns.barplot(x=needed_dataframe[col],y=needed_dataframe.index) 
                    plt.title('Barplot of {0}'.format(col))
                    plt.xlabel('{0}'.format(col))
                    plt.ylabel('Count of cars')
                    plt.savefig(newpath+'//barplot_{0}.png'.format(col))
                    plt.show()

                else:
                    plt.hist(needed_dataframe[col]) 
                    plt.title('Histogram of {0}'.format(col))
                    plt.xlabel('{0}'.format(col))
                    plt.ylabel('Frequecy of {0}'.format(col))
                    plt.savefig(newpath+'//histogram_{0}.png'.format(col))
                    plt.show()
                    plt.boxplot(needed_dataframe[col]) 
                    plt.title('Boxplot of {0}'.format(col))
                    plt.xlabel('{0}'.format(col))
                    plt.savefig(newpath+'//boxplot_{0}.png'.format(col))
                    plt.show()

```
